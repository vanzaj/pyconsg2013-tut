## Basic plotting

```python
dat = [3, -1, 0.5, 4, 2]
plot(dat)
```

```python
A = (0,1)
B = (1,0)
C = (2,1)

# x = [A[0], B[0], C[0]]
x = map(lambda i: i[0], [A,B,C])
y = [i[1] for i in [A,B,C]]

print(x,y)

plot(x, y, 'bo-')
```

```python
x = range(11)
y1 = [i**2 for i in x]
y2 = [i**2 for i in reversed(x)]

plot(x, y1, 'r-')
plot(x, y2, 'g-')
xlabel('x')
ylabel('y')
title('$f(x) \sim x^2$')
grid()
```
