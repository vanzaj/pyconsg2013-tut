```python
x = linspace(-1, 2, 20)
```

## figure and axis objects

```python
fig, ax = plt.subplots(figsize=(12,4))

ax.plot(x, x**2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('parabola')
```

## subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

ax[0].plot(x, x**2, 'r', x, (x-0.5)**3, 'b--')
ax[1].plot(x+1, exp(-(x+1)), 'g')
ax[1].set_xlabel('$t$')
ax[1].set_ylabel('$d$')
ax[1].set_title('$d(t) = e^{-t}$')
```
