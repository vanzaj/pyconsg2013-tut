```python
x = linspace(-1, 2, 20)
```

## explicit workflow

```python
# import matplotlib.pyplot as plt   # done by pylab

plt.figure(figsize=(9,4))
plt.plot(x, x**2, 'r-', linewidth=2)
plt.title('parabola', fontsize=16)
plt.show()
```

Default settings for fonts, colors, ticks, et al. are in stored in `matplotlib.rcParams`. Can be changed using something like:

    matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})



## figure and axis objects

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, x**2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('parabola')
ax.set_ylim(-1,5)
# fig.show()   # for non-inline backend
```

## subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))
ax[0].plot(x, x**2, 'r', x, (x-0.5)**3, 'b--')
ax[0].grid(True)
ax[1].plot(x+1, exp(-(x+1)), 'g')
ax[1].set_xlabel('$t$')
ax[1].text(0.7, 0.6, '$d(t) = e^{-t}$', fontsize=14, color='green')
```

## custom axes

```python
x = np.linspace(-0.75, 1., 100)
y = x**3
```

```python
fig, ax = subplots(1,2, figsize=(12,4))

# left
ax[0].spines['right'].set_color('none')
ax[0].spines['top'].set_color('none')

ax[0].xaxis.set_ticks_position('bottom')
ax[0].yaxis.set_ticks_position('left')

ax[0].plot(x, y);

# right
ax[1].spines['right'].set_color('none')
ax[1].spines['top'].set_color('none')

ax[1].xaxis.set_ticks_position('bottom')
ax[1].yaxis.set_ticks_position('left')

ax[1].spines['bottom'].set_position(('data',0)) # set position of x spine to x=0
ax[1].spines['left'].set_position(('data',0))   # set position of y spine to y=0

ax[1].plot(x, y);
```
