## Linear fit

```python
np = 50
x = linspace(-10, 20, np)

dat = (0.75 * x - 3) + randn(np)*2
plot(x, dat, 'ko')
```

```python
a, b = polyfit(x, dat, 1)
fit_line = polyval((a,b), x)

eq_lbl = '$y = %7.2f x + (%7.2f)$' % (a,b)

plot(x, dat, 'ko')
plot(x, fit_line,'b-', label=eq_lbl)
legend(loc='upper left')
```
